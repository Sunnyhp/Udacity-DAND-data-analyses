```python
import pandas as pd
import numpy as np

movies_data = pd.read_csv('movies.csv')
movies_data.head(2)
```

```python
movies_data.drop(['id', 'popularity', 'imdb_id', 'cast', 'homepage', 'director', 'tagline', 'overview', 'runtime', 'release_date', 'vote_count'], axis = 1, inplace = True)
movies_data.head(3)
```

```python
movies_data.isnull().sum()
```

```python
movies_data.dropna(inplace = True)
movies_data.isnull().sum()
```

```python
test_keyword = movies_data['keywords'].str.split('|').tolist()

keyword = []
for key in test_keyword:
    if 'based on novel' in key:
        keyword.append('Yes')
    else:
        keyword.append('No')  
```

```python
movies_data['based_on_novel'] = keyword
```

```python
movies_data.drop('keywords', axis = 'columns', inplace = True)
movies_data.head(5)
```

```python
split_genres = movies_data['genres'].str.split('|', expand = True).stack().reset_index(level=1, drop=True).rename('genres')
split_genres.head()
```

```python
split_production_companies = movies_data['production_companies'].str.split('|', expand = True).stack().reset_index(level=1, drop=True).rename('production_companies')
split_production_companies.head()
```

def convert_company(data):
    if 'Universal Pictures' == data:
        return 'Universal Pictures'
    elif 'Paramount Pictures' == data:
        return 'Paramount Pictures'
split_production_companies = split_production_companies.apply(convert_company)



```python
#split_production_companies
```

```python
movies_data = movies_data.drop('genres', axis = 1).join(split_genres)
```

```python
movies_data = movies_data.drop('production_companies', axis = 1).join(split_production_companies)
```

```python
movies_data.isnull().sum()
```

```python
movies_data.head(10)
```

```python
movies_data.dropna(inplace = True)
movies_data
```

```python
movies_data.to_csv('movies_cleaned.csv')
```
